```python
# Check if res is OK
res
```

```python
soup = BeautifulSoup(res, 'lxml') # BeautifulSoup(html, Parser Engine)
```

```python
# Let's see what we got
type(soup)
```

```python
article_list = soup.select(".rtddt")
```

```python
# Let's see what inside this list
article_list
```

```python
article_dict = {}
for article in article_list:
    article_dict['title'] = article.select('font')[0].text
    article_dict['pub_date'] = article.select('time')[0].text
    article_dict['type'] = article.select('h2')[0].text
    
    print "Title is {}".format(article.select('font')[0].text.encode('utf-8'))
    print "pub_date is {}".format(article.select('time')[0].text)
    print article.select('h2')[0].text
```

```python
article_dict
```

```python
import json
myjson = json.dumps(article_dict, ensure_ascii=False) # This is a unicode string
with open('articles.json', 'w') as f:
    f.write(myjson.encode('utf-8')) # encode our unicode string as 'utf-8' for writing out to file
```
