```python
# 列表頁爬蟲

import requests as r
from bs4 import BeautifulSoup

BASE_URL = "https://www.ptt.cc" 
URL = BASE_URL + "/bbs/Gossiping/index.html"
headers = {'cookie': "over18=1"}

res = r.get(URL, headers=headers)
soup = BeautifulSoup(res.text, 'lxml')

articles = soup.select('div.r-ent')

# 我們期望的每個Article的schema
article_dict = {
    'type': '',
    'title': '',
    'url': '',
    'score': ''
}

# =============article===============
# <div class="r-ent">
#      <div class="nrec"><span class="hl f2">1</span></div>
#      <div class="mark"></div>
#      <div class="title">
#          <a href="/bbs/Gossiping/M.1478225072.A.9AA.html">Re: [新聞] 惡鄰騷擾怎自保？持續紀錄分貝　報案檢舉</a>
#      </div>
#      <div class="meta">
#      <div class="date">11/04</div>
#      <div class="author">oooo</div>
# </div>
# ============article===================

for article in articles:
    article_dict['title'] = article.select_one('.title').text.strip()
    article_dict['type'] = article.select_one('.title').text.split('[')[1].split(']')[0]
    article_dict['url'] = BASE_URL + article.select_one('.title').select_one('a')['href']
    if article.select_one('span'):
        article_dict['score'] = article.select_one('span').text
```

```python
from peewee import *
from datetime import datetime

mysql_db = MySQLDatabase('ab104', host='localhost', user='root')

class BaseModel(Model):
    """A base model that will use our MySQL database"""
    class Meta:
        database = mysql_db

class News(BaseModel):
    id = IntegerField()
    title = CharField()
    url = CharField()
    content = TextField()
    pub_date = DateTimeField()
```

```python
first_news = News.create(title='nba', url='https://ianchenhq.com', content='do you watch nba?', pub_date = datetime.now())
```

```python
first_news.title = '籃球'
first_news.save()
```
